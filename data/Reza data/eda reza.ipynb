```python
import pandas as pd

# Load the dataset
df = pd.read_csv('final.csv')
```

```python
# Create a dictionary to map team names to their locations
team_locations = {}

# Populate the dictionary with team locations
for index, row in df.iterrows():
    home_team = row['home_team_name']
    location = row['location']
    if pd.notnull(home_team) and pd.notnull(location):
        team_locations[home_team] = location

team_locations['1. FC Heidenheim'] = 'Heidenheim'

# Function to fill missing locations based on the home team
def fill_location(row):
    return team_locations.get(row['home_team_name'], row['location'])

# Apply the function to the DataFrame
df['location'] = df.apply(fill_location, axis=1)

# Assuming '1. FC Heidenheim' has a unique and consistent team ID
heidenheim_id = 113  # Replace 113 with the actual team ID for '1. FC Heidenheim'

# Fill missing home_team_id and away_team_id for '1. FC Heidenheim'
df.loc[df['home_team_name'] == '1. FC Heidenheim', 'home_team_id'] = heidenheim_id
df.loc[df['away_team_name'] == '1. FC Heidenheim', 'away_team_id'] = heidenheim_id

# Save the updated DataFrame to a new CSV file
df.to_csv('updated_final.csv', index=False)

# Display the first few rows to verify
print("First few rows after filling missing locations and team IDs:")
display(df.head())
```
